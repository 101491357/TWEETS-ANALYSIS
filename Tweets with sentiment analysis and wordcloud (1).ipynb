```python
import nltk
import matplotlib.pyplot as plt
from collections import Counter


stopwords = nltk.corpus.stopwords.words('english')


with open(r"C:\\Users\\nanaa\\OneDrive\\Desktop\\Police have arrested one suspect in.txt", 'r') as file:
    text = file.read()


tokens = nltk.word_tokenize(text)
tokens = [word.lower() for word in tokens if word.isalpha()]  
tokens = [word for word in tokens if word not in stopwords]  


word_count = Counter(tokens)


N = 20
most_common_words = word_count.most_common(N)


words, counts = zip(*most_common_words)


plt.figure(figsize=(12, 6))
plt.bar(words, counts)
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Top {} Most Common Words".format(N))
plt.xticks(rotation=45)  
plt.show()
```

```python
import nltk
import matplotlib.pyplot as plt
from textblob import TextBlob
from collections import Counter


stopwords = nltk.corpus.stopwords.words('english')


with open(r"C:\\Users\\nanaa\\OneDrive\\Desktop\\Police have arrested one suspect in.txt", 'r') as file:
    text = file.read()


analysis = TextBlob(text)


polarity = analysis.sentiment.polarity
subjectivity = analysis.sentiment.subjectivity

print(f"Sentiment Polarity: {polarity}")
print(f"Sentiment Subjectivity: {subjectivity}")


tokens = nltk.word_tokenize(text)
tokens = [word.lower() for word in tokens if word.isalpha()] 
tokens = [word for word in tokens if word not in stopwords]  


word_count = Counter(tokens)


N = 20
most_common_words = word_count.most_common(N)


words, counts = zip(*most_common_words)


plt.figure(figsize=(12, 6))
plt.bar(words, counts)
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Top {} Most Common Words".format(N))
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer


nltk.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()


text = "Police have arrested one suspect in connection with a mass shooting at Morgan State University, while a second remains at large."

sentiment_scores = sid.polarity_scores(text)

if sentiment_scores['compound'] >= 0.05:
    sentiment = "positive"
elif sentiment_scores['compound'] <= -0.05:
    sentiment = "negative"
else:
    sentiment = "neutral"

print(f"Sentiment: {sentiment}")
```
